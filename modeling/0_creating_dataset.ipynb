```python
d = {"id":"", "label":"", "code":""}
id = []
label = []
code = []
with open("dataset/httpd_after_fix_extractor_0.pickle", mode = 'rb') as fp:
    while True:
        co = []
        try:
            item = pickle.load(fp)
            id.append(item['id'])
            label.append(item['label'])
            for i in item['functions'].values():
                co.append(i['code'])
            code.append(co)
        except EOFError:
            break
d['id'] = id
d['label'] = label
d['code'] = code
```

```python
httpd_after = pd.DataFrame(d)
httpd_after
```

## splits

```python
split = pd.read_csv('dataset/splits.csv')
split
```

```python
httpd_split = split.loc[split['project'] == 'httpd']
libav_split = split.loc[split['project'] == 'libav']
ffmpeg_split = split.loc[split['project'] == 'ffmpeg']
```

```python
split_temp = pd.concat([httpd_split, libav_split, ffmpeg_split])
split_temp
```

```python
split_temp.reset_index(inplace=True, drop=True)
split_temp
```

```python
final.reset_index(inplace=True, drop=True)
```

```python
final
```

```python
data = pd.merge(final, split_temp)
```

```python
data
```

```python
train = data[data['split'] == 'train']
train.reset_index(inplace=True, drop=True)
train.to_csv('train.csv')
```

```python
test = data[data['split'] == 'test']
test.reset_index(inplace=True, drop=True)
test.to_csv('test.csv')
```

```python
dev = data[data['split'] == 'dev']
dev.reset_index(inplace=True, drop=True)
dev.to_csv('dev.csv')
```
