```python
train = pd.read_csv('/content/drive/MyDrive/sanhak_d2a/train.csv')
train
```

```python
# 리스트 형태를 없애기 위해서 [, ] 삭제 +  /* */ 로 이루어져 있는 주석 제거
# \S -> 공백 문자 이외의 모든 문자, \s -> 공백 문자(빈칸, 탭 '\t', 개행 '\n')
train['noncomment_code'] = train['code'].map(lambda x : re.sub('\/\*[\S\s]*\*\/', '', x.replace('[', '').replace(']', '')))

# // 주석 해당 줄 모두 제거
train['noncomment_code'] = train['noncomment_code'].map(lambda x : re.sub('\/\/.*', '', x))

# 연속적인 \n 존재하면 한 개로 변경 
train['noncomment_code'] = train['noncomment_code'].map(lambda x :  re.sub('(\\\\n)+', '\\n', x))

train['noncomment_code'] = train['noncomment_code'].map(lambda x :  re.sub('\\\\t', '\\t', x))
train
```

```python
train.drop(['code', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
train.to_csv('/content/drive/MyDrive/sanhak_d2a/preprocessing_train.csv', index=False)
```
